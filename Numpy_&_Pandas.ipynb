```python
#2dimentional array - multinested array
my_list_1=[2,3,7,9,8]
my_list_2=[2,3,4,1,7]
my_list_3=[9,8,0,4,5]

arr_new=np.array([my_list_1,my_list_2,my_list_3])
```

```python
print(arr_new)
```

```python
arr_new.shape #3 rows and 5 columns
```

```python
arr_new.reshape(5,3) #transpose of a matrix (converting rows into columns and columns into rows)
```

```python
arr_new.reshape(4,5)
```

Numerical analysis

```python
a=np.array([2,3,5,6,8])
print("adding 1 to each element from the list",a+1)

print("subtract 1 from each element from the list",a-1)

print("multiply 5 to each element from the list",a*5)

print("square of each element from the list",a**2)
```

Generating random numbers

```python
np.arange(1,20,1) #from 1 to 20 we will get equally number (step size is 1)
```

```python
np.arange(1,20,3) #each number will have an equal difference of 3
```

```python
np.linspace(1,10,50)  #here 50 is the number of element we need between 1 to 10 (here 10 is not exclusive)
```

```python
np.linspace(1,10,100)
```

```python
#generating a random float number between 0 to 1 (here 1 wil be exclusive)
num=np.random.rand()
```

```python
print(num)
```

```python
int_num=np.random.randint(0,10)
print(int_num)
```

```python
np.random.rand(3,4) #here 3 means number of rows and 4 means the number of columns
```
